```python
from transformers import pipeline
from IPython.display import display
import ipywidgets as widgets
from transformers import pipeline  # Import the pipeline function

# Load the sentiment-analysis pipeline
sentiment_analysis = pipeline("sentiment-analysis")
```

```python
# Create a text input widget
text_input = widgets.Text(
    value='',
    placeholder='Type a sentence to analyze...',
    description='Input:',
    disabled=False
)

# Create a button to submit the text
submit_button = widgets.Button(
    description='Analyze Sentiment',
    disabled=False,
    button_style='success',
    tooltip='Click me',
    icon='check'
)

# Create an output widget to display the sentiment
output = widgets.Output()

# Define the button click event
def on_button_click(b):
    with output:
        output.clear_output()
        user_text = text_input.value
        if user_text.strip():
            result = sentiment_analysis(user_text)[0]
            sentiment = result['label']
            confidence = result['score']
            print(f"Sentiment: {sentiment}, Confidence: {confidence:.2f}")
        else:
            print("Please enter a valid sentence.")

# Bind the event to the button
submit_button.on_click(on_button_click)

# Display the widgets
display(text_input, submit_button, output)



```
